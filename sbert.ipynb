```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
nltk.download('stopwords')
import re
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.metrics.pairwise import cosine_similarity
nltk.download('punkt')
from sklearn.metrics.pairwise import euclidean_distances
```

```python
articles_df = pd.read_json('parsedText.json', lines=True)

documents = articles_df['text']
documents_df=pd.DataFrame(documents,columns=['text'])
```

```python
stop_words=stopwords.words('english')

extra = ['said','one','would','also','get','may','two','like','could']
words = []
for row in documents:
    row.strip()
    words.append(row)
for i in extra:
    stop_words.append(i)
filtered_list = []

test_list = ' '.join(words)

toeknized = word_tokenize(test_list)
for word in toeknized:
    if word.casefold() not in stop_words:
        filtered_list.append(word)

tfidfvectoriser=TfidfVectorizer()
tfidfvectoriser.fit(documents)
tfidf_vectors=tfidfvectoriser.transform(documents)
```

```python
pairwise_similarities=np.dot(tfidf_vectors,tfidf_vectors.T).toarray()
pairwise_differences=euclidean_distances(tfidf_vectors)
```

```python
def most_similar(doc_id,similarity_matrix,matrix):
    print (f'Document: {documents.iloc[doc_id]}')
    print ('\n')
    print ('Similar Documents:')
    if matrix=='Cosine Similarity':
        similar_ix=np.argsort(similarity_matrix[doc_id])[::-1]
    elif matrix=='Euclidean Distance':
        similar_ix=np.argsort(similarity_matrix[doc_id])
    for ix in similar_ix:
        if ix==doc_id:
            continue
        print('\n')
        print (f'Document: {documents.iloc[ix]}')
        print (f'{matrix} : {similarity_matrix[doc_id][ix]}')
```
